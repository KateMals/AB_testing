```python
all_v = df_v.REVENUE.value_counts().sum()
buy_v = df_v[df_v['REVENUE'] != 0].value_counts().sum()
all_v, buy_v
```

```python
all_c = df_c.REVENUE.value_counts().sum()
buy_c = df_c[df_c['REVENUE'] != 0].value_counts().sum()
all_c, buy_c
```

Вывод: 43 покупки в группе VARIANT и 47 покупок в группе CONTROL

```python
from statsmodels.stats import proportion
from scipy.stats import shapiro
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu
```

```python
chisq, pvalue, table = proportion.proportions_chisquare(np.array([buy_v, buy_c]),
np.array([all_v, all_c]))

print('Results are ','chisq =%.3f, pvalue = %.3f'%(chisq, pvalue))
```

p-value больше, чем 0,05 - выборки не имеют статистически значимых различий

```python
stats, pvalue = proportion.proportions_ztest(np.array([buy_v, buy_c]),
np.array([all_v, all_c]))

print('Results are ','z =%.3f, pvalue = %.3f'%(stats, pvalue))
```

p-value больше, чем 0,05 - выборки не имеют статистически значимых различий

```python
ttest_ind(df_v['REVENUE'], df_c['REVENUE'], equal_var=False)
```

При проведении Т-теста p-value больше, чем 0,05 - выборки не имеют статистически значимых различий
